# Subset Sum Problem

Given a set of non-negative integers and a target value `sum`, check if there is a subset of the given set whose sum is equal to the given `sum`.

We solve this using **Dynamic Programming** with 1D array space optimization.

```python
# Step 1: Base Case Setup
# Create a boolean DP array of size target_sum + 1, initialized to False.
# Set the base case dp[0] = True (an empty subset sums to 0).

def setup_dp(target_sum):
    dp = [False] * (target_sum + 1)
    dp[0] = True
    return dp

print("Initial DP State:", setup_dp(9))
```

### Step 2: The Backward Loop Transition

To avoid using the same element multiple times, we iterate through the sum backwards from `target_sum` down to the element's value `num`.

```python
# Step 2: Update the DP table with a subset of numbers
arr = [3, 4, 5]
target_sum = 9
dp = setup_dp(target_sum)

for num in arr:
    for j in range(target_sum, num - 1, -1):
        dp[j] = dp[j] or dp[j - num]
    print(f"After including {num:2}: {list(map(int, dp))}")

print(f"Is target sum {target_sum} possible?", dp[target_sum])
```

### Final Optimal Solution

Here is the complete optimal solution.

```python
def is_subset_sum(arr, target_sum):
    dp = [False] * (target_sum + 1)
    dp[0] = True
    
    for num in arr:
        for j in range(target_sum, num - 1, -1):
            dp[j] = dp[j] or dp[j - num]
            
    return dp[target_sum]

# Test cases
set_elements = [3, 34, 4, 12, 5, 2]
print("Subset Sum = 9:", is_subset_sum(set_elements, 9))
print("Subset Sum = 30:", is_subset_sum(set_elements, 30))
```
