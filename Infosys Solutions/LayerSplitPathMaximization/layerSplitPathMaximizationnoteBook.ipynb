# Layer-Split Path Maximization with Penalties
## Problem Description
Given an undirected graph where each node has a layer $L[u]$ and a value $V[u]$.
Find a simple path of non-decreasing layers $L[u_1] \le L[u_2] \le \dots \le L[u_t]$ that maximizes $\sum V[u] - \sum (L[u_{i+1}] - L[u_i])^2$.

## Example Test Case
Input: 3 nodes, 2 edges, $L = [1, 2, 3], V = [-5, 100, -10]$, edges $(0-1), (1-2)$.
Output: 100
Explanation: Starting at node 1 only gives value 100 with no penalties.

```python
def solve_layer_split_path(n, m, k, L, V, edges):
    adj = [[] for _ in range(n)]
    for u, v in edges:
        if L[u] <= L[v]: adj[u].append(v)
        if L[v] <= L[u]: adj[v].append(u)
        
    memo = {}
    
    def get_max(u, visited):
        state = (u, visited)
        if state in memo: return memo[state]
        max_val = 0
        for v in adj[u]:
            if L[v] == L[u]:
                if v not in visited:
                    val = V[v] + get_max(v, visited.union((v,)))
                    max_val = max(max_val, val)
            elif L[v] > L[u]:
                penalty = (L[v] - L[u]) ** 2
                val = V[v] - penalty + get_max(v, frozenset([v]))
                max_val = max(max_val, val)
        memo[state] = max_val
        return max_val
        
    ans = -float('inf')
    for start in range(n):
        ans = max(ans, V[start] + get_max(start, frozenset([start])))
    return ans

print(solve_layer_split_path(3, 2, 3, [1, 2, 3], [-5, 100, -10], [(0, 1), (1, 2)]))
```

### Key Idea: Visited Set Reset
Since we can only move to higher layers, once we leave a layer, we can never return to it. Thus, we only need to keep track of the visited set for the current layer, allowing us to reset the visited set whenever we transition to a strictly higher layer.
