# Lock & Parity
## Problem Description
You are given $N$ locks with values $L[i]$ and corresponding keys $L[j]$. Assign key $j$ to lock $i$ if $j < i$, $L[j] \neq L[i]$.
Find the minimum possible sum of effective values $|L[j] - L[i]|$ such that the number of even assignments is $\ge$ the number of odd assignments.

## Example Test Case
Input: $L = [41, 54, 15, 4, 54, 4]$
Output: 26
Explanation: We choose the pair $(1 \to 3)$ with cost $|41 - 15| = 26$ which is even.

```python
def get_min_valid_assignment(L):
    n = len(L)
    min_cost = float('inf')
    for i in range(n):
        for j in range(i):
            if L[j] != L[i] and (L[j] - L[i]) % 2 == 0:
                min_cost = min(min_cost, abs(L[j] - L[i]))
    return min_cost if min_cost != float('inf') else -1

print(get_min_valid_assignment([41, 54, 15, 4, 54, 4]))
```

### Why is a Single Even Edge Always Optimal?
Since every valid set of assignments must have `even >= odd` and size $\ge 1$, it must contain at least one even assignment.
Since all costs are positive, the cost of any valid set is at least the cost of its cheapest even assignment. A single even assignment is always valid, so the minimum cost is just the minimum cost of a single even edge.

```python
# Full solution implementation
L = [45, 6, 38, 6, 15, 38]
print(f"Result for L = {L}: {get_min_valid_assignment(L)}")
```
