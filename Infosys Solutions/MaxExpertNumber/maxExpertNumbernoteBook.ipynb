# Max Expert Number
## Problem Description
Given an array $A$ of skills. Partition $A$ into contiguous subarrays to maximize the sum of MEX values.

## Example Test Case
Input: $A = [0, 2, 1, 1]$
Output: 3
Explanation: Partition into $[0, 2, 1]$ (MEX = 3) and $[1]$ (MEX = 0).

```python
def max_expert_number(A):
    n = len(A)
    max_val = 1002
    last_pos = [-1] * (max_val + 2)
    dp = [0] * (n + 1)
    
    for i in range(1, n + 1):
        val = A[i-1]
        if val <= max_val:
            last_pos[val] = i - 1
            
        dp[i] = dp[i-1]
        curr_min = i - 1
        for m in range(1, max_val + 1):
            prev_last = last_pos[m-1]
            if prev_last < curr_min:
                curr_min = prev_last
            if curr_min >= 0:
                dp[i] = max(dp[i], dp[curr_min] + m)
            else:
                break
    return dp[n]

print(max_expert_number([0, 2, 1, 1]))
```

### Complexity Optimization
Since the maximum skill is $10^3$, the MEX of any subarray is at most $1002$. We can compute the largest starting position for each possible MEX $M$ at each step in $O(\max(A))$ time, leading to an overall time complexity of $O(N \cdot \max(A))$.
