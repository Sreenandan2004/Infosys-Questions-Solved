# Beautiful Subtree Set
## Problem Description
Given a colored tree. A beautiful set is a set of nodes of distinct colors such that every pair has an ancestor-descendant relationship.
Find the maximum size of a beautiful set in the subtree of $s$.

## Example Test Case
Input: $n=5, p=[0, 1, 2, 1, 3], color=[4, 3, 4, 3, 5]$, queries = $[4, 3, 3]$
Output: 5
Explanation: Sum of answers is $1 + 2 + 2 = 5$.

```python
def solve_beautiful_set(n, parents, colors, queries):
    children = [[] for _ in range(n + 1)]
    for i, p in enumerate(parents, 1):
        if p != 0:
            children[p].append(i)
            
    dp = [0] * (n + 1)
    
    def dfs_distinct(u, current_set):
        c = colors[u-1]
        already_present = c in current_set
        if not already_present:
            current_set.add(c)
        max_dist = len(current_set)
        for v in children[u]:
            max_dist = max(max_dist, dfs_distinct(v, current_set))
        if not already_present:
            current_set.remove(c)
        return max_dist
        
    for u in range(1, n + 1):
        dp[u] = dfs_distinct(u, set())
        
    ans = [0] * (n + 1)
    def compute_ans(u):
        max_val = dp[u]
        for v in children[u]:
            max_val = max(max_val, compute_ans(v))
        ans[u] = max_val
        return max_val
        
    compute_ans(1)
    return sum(ans[s] for s in queries) % (10**9 + 7)

print(solve_beautiful_set(5, [0, 1, 2, 1, 3], [4, 3, 4, 3, 5], [4, 3, 3]))
```

### Key Observation
Since any two nodes in a beautiful set must be ancestor-descendant, the entire set must lie on a single downward path. The problem reduces to finding the maximum number of distinct colors on any downward path in the subtree.
