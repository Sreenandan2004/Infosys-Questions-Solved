# Wine Bottle Transport - Quick Tutorial

Given an array `arr` representing $N$ houses built along a straight line. Each house has a certain number of wine bottles they want to sell (positive) or buy (negative). The total demand matches the total supply.

Transporting one bottle of wine to an adjacent house results in one unit of work.
Our goal is to find the **minimum work** required to fulfill all houses' demands.

### Example
- `arr` = `[5, -4, 1, -3, 1]`
- Output: `9`

```python
# Set up the input array
arr = [5, -4, 1, -3, 1]
n = len(arr)
print(f"Input array: {arr}")
print(f"Number of houses: {n}")
```

### Step 1: Understanding Prefix Imbalance

Since houses are arranged along a straight line, any surplus or deficit at the leftmost house must cross the boundary to the next house.
More generally, the net amount of wine that *must* cross the boundary between house `i` and house `i+1` to balance the left side is the cumulative sum of demands/supplies up to index `i`.

Let's calculate the running imbalance at each step.

```python
# Calculate the running imbalance (prefix sums)
imbalance = 0
imbalances = []
for x in arr:
    imbalance += x
    imbalances.append(imbalance)

print(f"Original Array:    {arr}")
print(f"Running Imbalance: {imbalances}")
```

### Step 2: Calculating Work and Final Optimal Solution

Since the houses are adjacent (distance = 1), transporting $k$ bottles across a boundary costs $k$ units of work.
Thus, the work crossing boundary `i` is simply `abs(imbalances[i])`.
By summing the absolute value of the running imbalances across all boundaries, we get the total work.

Let's implement the complete optimal solution with $O(N)$ time complexity and $O(1)$ space complexity.

```python
class Solution:
    def wineTrading(self, arr: list[int], n: int) -> int:
        work = 0
        imbalance = 0
        for bottles in arr:
            imbalance += bottles
            work += abs(imbalance)
        return work

# Instantiate and test the solution
sol = Solution()
test_cases = [
    [5, -4, 1, -3, 1],
    [-1000, -1000, -1000, 1000, 1000, 1000]
]

for case in test_cases:
    print(f"Input: {case} -> Minimum Work: {sol.wineTrading(case, len(case))}")
```
