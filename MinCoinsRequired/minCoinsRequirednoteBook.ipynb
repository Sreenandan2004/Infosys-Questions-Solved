# Minimum Coins Required
## Problem Description
Given an array `coins` representing coin denominations and a target value `V`, find the minimum number of coins required to make the target value `V`. If it is not possible to make the target value, return `-1`.
## Example Test Case
- Input: `coins = [25, 10, 5]`, `V = 30`
- Output: `2` (using one 25-cent coin and one 5-cent coin)

```python
# Step 1: Initialize the DP array
# We set the size of the DP array to V + 1.
# All elements are initialized to infinity except the base case dp[0] = 0.

def initialize_dp(V):
    dp = [float('inf')] * (V + 1)
    dp[0] = 0
    return dp

print("Initialized DP for V = 11:", initialize_dp(11))
```

### Step 2: Recurrence Relation and State Transitions

For each value from `1` to `V`, we try to reduce the target using every coin denomination. The recurrence relation is:
$$dp[i] = \min_{c \in coins} (dp[i-c] + 1)$$

Here is how we transition step by step.

```python
# Step 2: Loop and state updates simulation
coins = [1, 5, 6, 9]
V = 11
dp = initialize_dp(V)

for i in range(1, V + 1):
    for coin in coins:
        if i - coin >= 0:
            dp[i] = min(dp[i], dp[i - coin] + 1)

print(f"Fully filled DP table for V = {V}:", dp)
print(f"Minimum coins needed for V = {V}:", dp[V] if dp[V] != float('inf') else -1)
```

### Final Optimal Solution

Here is the complete optimal solution packaged as a clean, reusable function.

```python
def min_coins(coins, V):
    dp = [float('inf')] * (V + 1)
    dp[0] = 0
    
    for i in range(1, V + 1):
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i], dp[i - coin] + 1)
                
    return dp[V] if dp[V] != float('inf') else -1

# Test cases
print("Coins [25, 10, 5], V = 30:", min_coins([25, 10, 5], 30))
print("Coins [9, 6, 5, 1], V = 11:", min_coins([9, 6, 5, 1], 11))
```
