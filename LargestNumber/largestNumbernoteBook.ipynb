# Largest Number

Given an array of strings representing non-negative integers, arrange them such that they form the largest possible number when concatenated.

We solve this using a **Greedy Sorting** approach with a custom comparator.

```python
# Step 1: Defining the Custom Comparator
# In Python 3, we use functools.cmp_to_key to use a classic compare(x, y) function in sort.

import functools

def compare(x, y):
    if x + y > y + x:
        return -1  # x should precede y
    elif x + y < y + x:
        return 1   # y should precede x
    return 0

print("Comparing '3' and '30':", compare('3', '30'))
```

### Step 2: Sorting and Handling Edge Cases

If the array consists of only zeros, the sorted result will be `['0', '0']`. Concatenating them directly yields `'00'`, which is mathematically just `'0'`. We must handle this.

```python
# Step 2: Sort the array and format the result
arr = ["3", "30", "34", "5", "9"]
sorted_arr = sorted(arr, key=functools.cmp_to_key(compare))

print("Sorted array:", sorted_arr)

# Handle edge case
result = "0" if sorted_arr[0] == "0" else "".join(sorted_arr)
print("Largest Number Result:", result)
```

### Final Optimal Solution

Here is the complete optimal solution packaged together.

```python
import functools

def largest_number(arr):
    def compare(x, y):
        if x + y > y + x:
            return -1
        elif x + y < y + x:
            return 1
        return 0

    sorted_arr = sorted(arr, key=functools.cmp_to_key(compare))
    
    if sorted_arr[0] == "0":
        return "0"
        
    return "".join(sorted_arr)

# Test case
print("Result:", largest_number(["54", "546", "548", "60"]))
```
